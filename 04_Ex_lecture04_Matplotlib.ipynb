Create a square figure.

Plot the $\log(x)$ function between (1, 100). Use a red dotted line with width = 2.5 points.

Set the x and y ranges.

Add the x and y labels.

```python
import numpy as np
import matplotlib.pyplot as plt
n_points=20
x_min=1.
x_max=100.
dist_x=(x_max-x_min)

fig,ax=plt.subplots(figsize=(10,10))
x=np.linspace(x_min,x_max,n_points)
y1=np.log(x)
y2=x**0.3
temp=np.concatenate((y1,y2))
dist_y=np.max(temp)-np.min(temp)
ax.plot(x,y1,c='red', linestyle=':', lw=2.5, label=r"$\log(x) $")
ax.plot(x,y2,c='cyan', linestyle='-',marker='*',
        markerfacecolor='blue', markeredgecolor='white',
        markersize=10, lw=1.5,label=r'$x^{0.3}$')
ax.legend(fontsize=15)
ax.set_xlim(x_min,x_max)
ax.set_ylim(np.min(temp),np.max(temp))
ax.set_title('my plots',fontsize=15,fontstyle='italic',c='violet')

#done in 23/03 lecture -> correction is from min 153
```

## Exercise 2


Plot a figure with two subplots (2 rows, one column).

In the top graph, plot the functions $\sin(x^2)$ and $\cos(x^2)$ between $[0, 2\pi]$. 

In the bottom graph, plot $\sin(x^2)\cos(x^2)$ in the same interval.

Add labels to the figure and to the axes. Adjust the spacing between subplots.

Save the figure as a pdf file.

```python
import numpy as np
import matplotlib.pyplot as plt

n_points=200
x_min=0
x_max=2*np.pi
x=np.linspace(x_min,x_max,n_points)
y1=np.sin(x**2)
y2=np.cos(x**2)
y3=y1*y2

fig,ax=plt.subplots(2,1,figsize=(10,10))

# subplot 0
ax[0].plot(x, y1, c='red', linestyle=':', lw=2, 
           label=r"$\sin(x^2) $")
ax[0].plot(x, y2, c='blue', linestyle='-', lw=2, 
           label=r"$\cos(x^2) $")
ax[0].set_xlabel('x',fontsize=20)
ax[0].set_ylabel('y',fontsize=20)

# subplot 1
ax[0].legend(fontsize=15)
ax[1].plot(x, y3, c='cyan', linestyle='-.',  lw=2, 
           label=r'\sin(x^2) \codt \cos(x^2)')
ax[1].legend(fontsize=15)
ax[1].set_xlabel('x',fontsize=20)
ax[1].set_ylabel('y',fontsize=20)


# the x axis label of the top plot could disappear underneath
# the bottom plot: in such case you can use the following function
plt.subplots_adjust(hspace=0.25)

# in case you want to save the figure, just type:
# fig.savefig('sin_cos_x2.pdf')
# the file with the figure is saved in the current directory,
# which can be obtained through the use of
%pwd
```
